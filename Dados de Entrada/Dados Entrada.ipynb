```python
import numpy as np

body_acc_x_test = np.loadtxt("HAR Dataset NOVO/UCI HAR Dataset/UCI HAR Dataset/test/Inertial Signals/body_acc_x_test.txt")[0]
body_acc_y_test = np.loadtxt("HAR Dataset NOVO/UCI HAR Dataset/UCI HAR Dataset/test/Inertial Signals/body_acc_y_test.txt")[0]
body_acc_z_test = np.loadtxt("HAR Dataset NOVO/UCI HAR Dataset/UCI HAR Dataset/test/Inertial Signals/body_acc_z_test.txt")[0]
body_gyro_x_test = np.loadtxt("HAR Dataset NOVO/UCI HAR Dataset/UCI HAR Dataset/test/Inertial Signals/body_gyro_x_test.txt")[0]
body_gyro_y_test = np.loadtxt("HAR Dataset NOVO/UCI HAR Dataset/UCI HAR Dataset/test/Inertial Signals/body_gyro_y_test.txt")[0]
body_gyro_z_test = np.loadtxt("HAR Dataset NOVO/UCI HAR Dataset/UCI HAR Dataset/test/Inertial Signals/body_gyro_z_test.txt")[0]
total_acc_x_test = np.loadtxt("HAR Dataset NOVO/UCI HAR Dataset/UCI HAR Dataset/test/Inertial Signals/total_acc_x_test.txt")[0]
total_acc_y_test = np.loadtxt("HAR Dataset NOVO/UCI HAR Dataset/UCI HAR Dataset/test/Inertial Signals/total_acc_y_test.txt")[0]
total_acc_z_test = np.loadtxt("HAR Dataset NOVO/UCI HAR Dataset/UCI HAR Dataset/test/Inertial Signals/total_acc_z_test.txt")[0]
tempo = np.arange(0, 200)

body_acc_x_test, tempo
```

```python
import matplotlib.pyplot as plt

plt.plot(body_acc_x_test)
plt.plot(body_acc_y_test)
plt.plot(body_acc_z_test)
plt.grid()
plt.legend(['Acelerao em $x$', 'Acelerao em $y$', 'Acelerao em $z$'])
plt.title('Sinais de Entrada Relativos  Acelerao')
plt.xlabel('Tempo')
plt.ylabel('Acelerao')
plt.savefig("Imagens Simulao/Sinal Entrada Acelerao")
```

```python
import matplotlib.pyplot as plt

plt.plot(body_gyro_x_test)
plt.plot(body_gyro_y_test)
plt.plot(body_gyro_z_test)
plt.grid()
plt.legend(['Inrcia em $x$', 'Inrcia em $y$', 'Inrcia em $z$'])
plt.title('Sinais de Entrada do Giroscpio')
plt.xlabel('Tempo')
plt.ylabel('Inrcia')
plt.savefig("Imagens Simulao/Sinal Entrada Inrcia")
```
