```python
import pandas as pd
```

```python
y_train = pd.read_csv("../Dataset/UCI HAR Dataset/UCI HAR Dataset/train/y_train.txt", header=None, sep="\s+")

labels = pd.read_csv("../Dataset/UCI HAR Dataset/UCI HAR Dataset/activity_labels.txt", header=None, sep="\s+")

y_train = pd.merge(y_train, labels)[1]

y_train
```

```python
import matplotlib.pyplot as plt

plt.hist(y_train, bins=11)
plt.title("Quantidade de Amostras Por Movimento")
plt.xlabel("Movimento")
plt.ylabel("Quantidade")
plt.xticks(rotation="vertical")
plt.savefig("Distribuição Atividades Dataset Saída.png", bbox_inches='tight')
```
